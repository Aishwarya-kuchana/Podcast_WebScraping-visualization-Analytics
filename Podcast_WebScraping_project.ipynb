```python
import time
import random
full_podcasts = []
headers = {"User-Agent": "Mozilla/5.0"}
for genre_name, genre_url in zip(podcast_genre,podcast_genre_url):
    print(f"\nScraping genre: {genre_name} ({genre_url})")
    
    try:
        r = requests.get(genre_url, headers=headers, timeout=15)
        g_soup = BeautifulSoup(r.text, "html.parser")
    except Exception as e:
        print(f"❌ Skipped {genre_name} due to {e}")
        continue

    # Each section (Top Shows, Essentials, etc.)
    sections = g_soup.find_all("h2", class_="title svelte-rnrb59 title-link")

    for section in sections:
        section_title = section.find("h2")
        section_name = section_title.text.strip() if section_title else "Unnamed Section"

        # Find all podcast links under this section
        podcasts_list = section.find_all("span", class_="product-lockup__title link svelte-1tr6noh")

        for p in podcasts_list:
            podcast_name = p.text.strip()
            podcast_link = p.get("href")

            full_podcasts.append({
                "Podcast_genre": genre_name,
                "Podcast_genre_url": genre_url,
                "Genre_Section": section_name,
                "Podcast_name": podcast_name,
                "Podcast_url": podcast_link
            })
    
    # polite pause to avoid being blocked
    time.sleep(random.uniform(1, 2))

print(f"\n✅ Total podcasts scraped: {len(full_podcasts)}")


# -------------------------------------------------
# Step 3: Convert to DataFrame and save
# -------------------------------------------------

df = pd.DataFrame(full_podcasts)
df.to_csv("apple_podcasts_clean.csv", index=False, encoding="utf-8-sig")

print("\n✅ Data saved to apple_podcasts_clean.csv")
```

```python
import requests

url = "https://itunes.apple.com/WebObjects/MZStoreServices.woa/ws/genres"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
genres_data = response.json()

# Main Podcasts genre
podcasts_genre = genres_data['26']

# Subgenres under Podcasts
podcast_subgenres = podcasts_genre.get('subgenres', {})

podcast_genres_list = []
for g_id, g_info in podcast_subgenres.items():
    podcast_genres_list.append({
        'name': g_info['name'],  # e.g., Business
        'id': g_info['id']       # e.g., 1321
    })

print(podcast_genres_list)
```

```python
import requests

headers = {"User-Agent": "Mozilla/5.0"}

full_podcasts = []

for genre in podcast_genres_list:  # podcast_genres_list = [{'name': 'Business', 'id':'1321'}, ...]
    genre_name = genre['name']
    genre_id = genre['id']
    
    # JSON URL for top podcasts
    url = f"https://itunes.apple.com/in/rss/toppodcasts/limit=100/genre={genre_id}/json"
    
    response = requests.get(url, headers=headers)
    data = response.json()
    
    if 'feed' in data and 'entry' in data['feed']:
        for p in data['feed']['entry']:
            full_podcasts.append({
                'Main_genre': genre_name,                       # Main category
                'Genre_id': genre_id,                           # Genre ID
                'Subcategory': p['category']['attributes']['term'], # Section within genre
                'Podcast_name': p['im:name']['label'],
                'Creator': p['im:artist']['label'],
                'Description': p['summary']['label'],
                'Podcast_url': p['link']['attributes']['href']
            })
```

```python
import requests
from bs4 import BeautifulSoup
import time
import random

# Make sure you have your all_podcasts list from JSON feed
# Example: all_podcasts = [{'Podcast_name':..., 'Podcast_url':..., ...}, ...]

headers = {"User-Agent": "Mozilla/5.0"}

for i, podcast in enumerate(full_podcasts, start=1):
    url = podcast['Podcast_url']
    
    try:
        response = requests.get(url, headers=headers, timeout=10)
        soup = BeautifulSoup(response.text, "html.parser")
        
        # --- Number of episodes ---
        episodes_tag = soup.find("div", class_="content svelte-sy8mrl")
        num_episodes = episodes_tag.text.strip() if episodes_tag else None
        
        # --- Rating ---
        rating_tag = soup.find("div", class_="stats__main svelte-1rwyjxh")
        rating = rating_tag.text.strip() if rating_tag else None

        #content_tag = soup.find("li", class_="content svelte-sy8mrl")
        #content_rating = content_tag.text.strip() if content_tag else None
        
        # Add to podcast dict
        podcast['episodes'] = num_episodes
        podcast['Rating'] = rating
        #Podcast["Content_Rating"]=content_rating
        
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        podcast['Num_episodes'] = None
        podcast['Rating'] = None
    
    # Optional: progress display
    if i % 50 == 0:
        print(f"Processed {i}/{len(full_podcasts)} podcasts...")
    
    # Be polite to Apple servers
    time.sleep(random.uniform(1, 2))
```

```python
import requests
from bs4 import BeautifulSoup
import time
import random
import pandas as pd

headers = {"User-Agent": "Mozilla/5.0"}

for i, podcast in enumerate(full_podcasts, start=1):
    url = podcast["Podcast_url"]

    try:
        response = requests.get(url, headers=headers, timeout=15)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")

        episodes = None
        content_rating = None

        # Loop through all <li> tags containing title and content
        for li in soup.find_all("li"):
            title_tag = li.find("div", class_="title svelte-sy8mrl")
            content_tag = li.find("div", class_="content svelte-sy8mrl")

            if title_tag and content_tag:
                title_text = title_tag.text.strip().lower()

                if title_text == "episodes":
                    episodes = content_tag.text.strip()
                elif title_text == "content rating":
                    content_rating = content_tag.text.strip()

        podcast["Num_episodes"] = episodes
        podcast["Content_rating"] = content_rating

    except Exception as e:
        print(f"Error fetching {url}: {e}")
        podcast["Num_episodes"] = None
        podcast["Content_rating"] = None

    # Progress every 50 podcasts
    if i % 50 == 0:
        print(f"Processed {i}/{len(full_podcasts)} podcasts...")

    # Pause between requests to avoid blocking
    time.sleep(random.uniform(1, 2))

print("✅ Scraping complete!")

```

```python
def convert_num_episodes(x):
    if pd.isna(x):
        return None
    x = str(x).lower().replace(',', '').strip()
    if 'k' in x:
        return float(x.replace('k','')) * 1000
    elif 'm' in x:
        return float(x.replace('m','')) * 1000000
    else:
        return float(x)
```

```python
df.loc[:, "Num_episodes"] = df["Num_episodes"].apply(convert_num_episodes).astype(int)
```
